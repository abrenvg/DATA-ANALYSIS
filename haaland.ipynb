```python
hal['X']=hal['X']*100
hal['Y']=hal['Y']*100
hal.head()
```

```python
hal['shotType'].unique()
```

```python
totshot=hal.shape[0]
right=hal[hal['shotType']=='RightFoot'].shape[0]
left=hal[hal['shotType']=='LeftFoot'].shape[0]
head=hal[hal['shotType']=='Head'].shape[0]
other=hal[hal['shotType']=='OtherBodyPart'].shape[0]
totgol=hal[hal['result']=='Goal'].shape[0]
totxg=hal['xG'].sum()
xGpershot=totxg/totshot
points_avg_dist=hal['X'].mean()
actual_avg_dist=120-(hal['X']*1.2).mean()#as  normal pitch len is 120 yards
```

```python
import matplotlib.font_manager as font_manager
font_path="C:/Users/abren/Downloads/Arvo-Regular.ttf"
font=font_manager.FontProperties(fname=font_path)
```

```python
fig=plt.figure(figsize=(8,12))
fig.patch.set_facecolor(bg)

ax1=fig.add_axes([0,.7,1,.2])# create a grid 1st no-leftcorner,2nd-bottom corner,3-width,4th-heigh
ax1.set_facecolor(bg)
ax1.set_xlim(0,1)
ax1.set_ylim(0,1)

ax1.text(x=.5,y=.85,s='ERLING HAALAND',fontsize=20,fontproperties=font,fontweight='bold',color='white',ha='center')#ha-horizontal alighnment)
ax1.text(x=.5,y=.75,s='ALL NON-PENALTY SHOTS PL 22/23',fontsize=14,fontproperties=font,fontweight='bold',color='white',ha='center')#ha-horizontal alighnment)
ax1.text(x=.25,y=.5,s='LOW QUALITY CHANCE',fontsize=12,fontproperties=font,fontweight='bold',color='white',ha='center')#ha-horizontal alighnment)

ax1.scatter(x=.4,y=.53,s=150,color=bg,edgecolor='white',linewidth=.8)
ax1.scatter(x=.43,y=.53,s=200,color=bg,edgecolor='white',linewidth=.8)
ax1.scatter(x=.47,y=.53,s=250,color=bg,edgecolor='white',linewidth=.8)
ax1.scatter(x=.51,y=.53,s=300,color=bg,edgecolor='white',linewidth=.8)
ax1.scatter(x=.55,y=.53,s=350,color=bg,edgecolor='white',linewidth=.8)

ax1.text(x=.71,y=.5,s='HIGH QUALITY CHANCE',fontsize=12,fontproperties=font,fontweight='bold',color='white',ha='center')#ha-horizontal alighnment)

ax1.text(x=.39,y=.38,s='GOAL',fontsize=12,fontproperties=font,fontweight='bold',color='white',ha='right')#ha-horizontal alighnment)
ax1.scatter(x=.42,y=.4,s=100,color='red',edgecolor='white',linewidth=.8,alpha=.7)#alpha for transparency
ax1.scatter(x=.5,y=.4,s=100,color=bg,edgecolor='white',linewidth=.8,alpha=.7)#alpha for transparency
ax1.text(x=.52,y=.38,s='NO GOAL',fontsize=12,fontproperties=font,fontweight='bold',color='white',ha='left')#ha-horizontal alighnment)



ax2=fig.add_axes([0.05,.25,.9,.5])
ax2.set_facecolor(bg)

pitch=VerticalPitch(pitch_type='opta',half=True,pitch_color=bg,pad_bottom=.5,line_color='white',linewidth=.75,axis=True,label=True)
pitch.draw(ax=ax2)

ax2.scatter(x=90,y=points_avg_dist,s=100,color='white',linewidth=.8)
ax2.plot([90,90],[100,points_avg_dist],color='white',linewidth=2)
ax2.text(x=89.2,y=points_avg_dist-5,s=f'AVERAGE DISTANCE\n{actual_avg_dist:.1f}yards',fontsize=9,fontproperties=font,color='white',ha='center')

for s in hal.to_dict(orient='records'):#to loop through dictionary
    pitch.scatter(s['X'],s['Y'],s=300*s['xG'],color='red' if s['result']=='Goal' else bg,ax=ax2,alpha=.7,linewidth=.8,edgecolor='white')#s-size-scaled based on xG

ax3=fig.add_axes([0,.2,1,.05])
ax3.set_facecolor(bg)

ax3.text(x=.25,y=.5,s='SHOTS',fontsize=10,fontproperties=font,fontweight='bold',color='white',ha='center')
ax3.text(x=.25,y=0,s=f'{totshot}',fontsize=14,fontproperties=font,fontweight='bold',color='red',ha='center')


ax3.text(x=.38,y=.5,s='GOALS',fontsize=10,fontproperties=font,fontweight='bold',color='white',ha='center')
ax3.text(x=.38,y=0,s=f'{totgol}',fontsize=14,fontproperties=font,fontweight='bold',color='red',ha='center')


ax3.text(x=.50,y=.5,s='TOTAL xG',fontsize=10,fontproperties=font,fontweight='bold',color='white',ha='center')
ax3.text(x=.50,y=0,s=f'{totxg:.2f}',fontsize=14,fontproperties=font,fontweight='bold',color='red',ha='center')#.2f-to scale upto 2 decimal


ax3.text(x=.63,y=.5,s='xG/SHOT',fontsize=10,fontproperties=font,fontweight='bold',color='white',ha='center')
ax3.text(x=.63,y=0,s=f'{xGpershot:.2f}',fontsize=14,fontproperties=font,fontweight='bold',color='red',ha='center')









```
